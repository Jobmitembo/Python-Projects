```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Python Data/world_population.csv')

df
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)   ### This changes display of scientific figures to understandable floats
```

```python
### Display info abt the dataset (dataframe)
df.info()
```

```python
### See a high level summary of your dataframe
df.describe()
```

```python
### Gives us sum of null values
df.isnull().sum()
```

```python
##### Sort desc according to the selected column, head() prints out the top5

df.sort_values(by="2022 Population", ascending = False) .head()
```

```python
# Establish the correlation between the population and the rest of the variables
df.corr()
```

```python
### We then visualize these corellations

sns.heatmap(df.corr(), annot = True)
plt.rcParams['figure.figsize'] = (10,7)
plt.show()
```

```python
#Calculate average population YoY per continent

df.groupby('Continent').mean()
```

```python
#Dive into  "Oceania" continent, I am curious to findout which countries make up this continent and how their population looks like

df[df['Continent'].str.contains('Oceania')]
```

```python
#Reorganise this data for easy visualization in order to see the trend across the different years

df2 = df.groupby('Continent')['1970 Population',
 '1980 Population',
 '1990 Population',
 '2000 Population',
 '2010 Population',
 '2015 Population',
 '2020 Population',
'2022 Population'].mean().sort_values(by ='2022 Population', ascending = False)
df2
```

```python

#Create a new dataFrame where we transpose this data in order for it to be easy to visualize
df3 = df2.transpose()
df3
```

```python
#Visualize the trend

plt.rcParams['figure.figsize']=(15,4)
df3.plot()
plt.xlabel('Year')
plt.ylabel('Mean population')
plt.title('World Pupolation')
#plt.ylim(bottom =0)
plt.show()
```
