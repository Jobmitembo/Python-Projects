```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Python Data/Customer Call List.xlsx')
df
```

```python
###Remove duplicates

df = df.drop_duplicates()
df
```

```python
#### Remove useless columns

df = df.drop(columns = 'Not_Useful_Column')
df
```

```python
### Clean Last_name column to remove the unnecessary characters

df['Last_Name']= df['Last_Name'].str.strip('/._') #If u notice we have put all the characters we want to trim in the same bracket!
df
```

```python
### Now we are cleaning the Phone Number column to be uniform (Currently the phone numbers aren't of the same format)

###We shall start by making all phone numbers drop non alphabetic or numerical characters as done below:

df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','')  ### What this will do is it will match all the rest of the characters except those mentiones above, then all the matched characters will be replaced by nothing. This puts in some regex knowledge
df
```

```python
### Now that the phone Number column includes both numbers(floats) and strings (NaN), we nee to first convert the whole column to string
### We can do this using for loop or lamba, we are chosing lambda here

df['Phone_Number']= df['Phone_Number'].apply(lambda x: str(x))  ### This will change change every Phone number column entry to a string

### Now we we use lambda again to add hiphens. we shall use the index because we want the phone number to look like this (123-456-7890)

df['Phone_Number']= df['Phone_Number'].apply(lambda x: x[0:3]+ '-' + x[3:6]+'-'+ x[6:10])
df
```

```python
### Now after that, we are going to remove the "nan--" AND "nA--", we shall just replace these with nothing

df['Phone_Number'] = df['Phone_Number'].str.replace('nan--','')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--','')

### Now we are done cleaning the Phone number column and we can print our dataframe

df
```

```python
### Now we gotta split the address and make three different columns out of it and name these columns as well

df['Address'].str.split(',',2, expand = True) ### Note that the new columns are not named, we name them below:
df[['Street_address','State','Zip_code']]= df['Address'].str.split(',',2, expand = True) ### Done by creating a list of lists
df
```

```python
### Now Replace all Yes with "Y" and NO with "N"
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')
df
```

```python
### Remove N/A, Nan,and None. All those are technically nulls so we cannot replace them, we shall use another function called fillna() to replace all those with blank space

df= df.fillna('')
df
```

```python
### Now we are returning only those who have a phone number and do not contact is "N" meaning they should be contacted, we shall use a For loop. This is done
### to make sure the list doesn't include those who said we should not contact them

for x in df.index:
  if df.loc[x,'Do_Not_Contact']== 'Y':
    df.drop(x, inplace = True)   ### inplace helps us to replace this in the original dataframe


for x in df.index:
  if df.loc[x,'Phone_Number']== '':
    df.drop(x, inplace = True)   ### inplace helps us to replace this in the original dataframe
df
```

```python
### Now we reset the index to avoid missing values

df.reset_index(drop = True) ##We add drop = True to drop the disorganised index, otherwise we would have the new index and the disorganised one
```
